```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('heart.csv')

corr_matrix = df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.show()
```

```python

plt.figure(figsize=(10,5))
custom_palette = ["#0000ff", "#ff00ff"]  
sns.countplot(data=df, x='output', palette=custom_palette, hue='sex')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='sex', palette=custom_palette, hue='sex')
plt.show()
```

# Large Discrepancy between count of males and females in the dataset. Possible subject for preprocessing.

```python
male_df = df[df['sex'] == 0]

plt.figure(figsize=(10, 5))
custom_palette = ["#0000ff", "#0000ff"]  
sns.countplot(data=male_df, x='output', palette=custom_palette)
plt.show()
```

```python
import numpy as np

# Identify indices of males with output 1
male_indices_with_output_1 = df[(df['sex'] == 1) & (df['output'] == 1)].index

# Calculate the number of males to remove (66% of males with output 1)
num_males_to_remove = int(0.66 * len(male_indices_with_output_1))

# Randomly select indices to remove
indices_to_remove = np.random.choice(male_indices_with_output_1, num_males_to_remove, replace=False)

# Drop selected indices from DataFrame
df = df.drop(indices_to_remove)
```

```python
male_df = df[df['sex'] == 0]

plt.figure(figsize=(10, 5))
custom_palette = ["#0000ff", "#0000ff"]  
sns.countplot(data=male_df, x='output', palette=custom_palette)
plt.show()
```
